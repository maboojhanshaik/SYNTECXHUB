Importing the required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully.")
```

Data Loading And Cleaning 

```python
matches = pd.read_csv('ipl_matches_data.csv')
deliveries = pd.read_csv('ball_by_ball_data.csv')

pd.set_option('display.max_columns', None)

display(matches.head())
display(deliveries.head())

display(matches.describe().round(2))
display(deliveries.describe().round(2))
```

Mergening the data sets

```python
df = pd.merge(deliveries,matches[['match_id', 'season_id']], on='match_id', how='left')
display(df.head())
display(df.describe().round(2))
```

Top Scorers and Strike Rates

```python
#Calculateing total runs per batter
batter_stats = df.groupby('batter').agg({
    'batter_runs': 'sum',
    'is_wide_ball': 'count' 
}).reset_index()
batter_stats
```

```python
# Calculateing valid balls (excluding wide balls for strike rate calculation
valid_balls = df[df['is_wide_ball'] == False].groupby('batter').size().reset_index(name='balls_faced')
batter_stats = pd.merge(batter_stats, valid_balls, on='batter', how='left')
batter_stats
```

```python
# Calculateing Strike Rate: (Runs / Balls Faced) * 100
batter_stats['strike_rate'] = ((batter_stats['batter_runs'] / batter_stats['balls_faced']) * 100).round(2)
batter_stats
```

```python
# Filtering  for top 10 scorers
top_10_scorers = batter_stats.sort_values(by='batter_runs', ascending=False).head(10)
top_10_scorers
```

```python
#plotting the graph
ax=sns.barplot(data=top_10_scorers, x='batter_runs', y='batter', palette='viridis')
for container in ax.containers:
    ax.bar_label(container, padding=3)
plt.title('Top 10 All-Time IPL Scorers', fontsize=16)
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.savefig('Top 10 all- time ipl scorers.jpeg');
plt.show()
```

```python
print("Top 5 Scorers with Strike Rates:")
print(top_10_scorers[['batter', 'batter_runs', 'strike_rate']].head())
```

```python
# Calculateing matches played by each team
team1_counts = matches['team1'].value_counts()
print(team1_counts)
team2_counts = matches['team2'].value_counts()
print(team2_counts)
total_matches = team1_counts.add(team2_counts, fill_value=0)
total_matches
```

```python
# Calculateing wins
wins = matches['match_winner'].value_counts()
wins
```

```python
# Calculate Win Rate Percentage
win_rate = (wins / total_matches * 100).sort_values(ascending=False).dropna().round(2)
win_rate
```

```python
#ploting the graph
ax=sns.barplot(x=win_rate.index, y=win_rate.values, palette='magma', order=win_rate.index)
for container in ax.containers:
    ax.bar_label(container, padding=3,fmt='%.1f%%')
plt.title('Team Win Rates (%)', fontsize=16)
plt.ylabel('Win Percentage')
plt.xlabel('Team ID')
plt.xticks(rotation=45)
plt.savefig('Team Win Rate (%).jpeg');
plt.show()
```

```python
# Calculateing total runs scored in each season
seasonal_runs = df.groupby('season_id_x')['total_runs'].sum().reset_index()
seasonal_runs
```

```python
#plotting the graph
sns.lineplot(data=seasonal_runs, x='season_id_x', y='total_runs', marker='o', color='crimson', linewidth=2.5)
plt.title('IPL Total Runs Evolution (Season by Season)', fontsize=16)
for i in range(seasonal_runs.shape[0]):
    plt.text(seasonal_runs.season_id_x.iloc[i], 
             seasonal_runs.total_runs.iloc[i] + 100, 
             f'{int(seasonal_runs.total_runs.iloc[i]):,}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.ylabel('Total Runs')
plt.xlabel('Season')
plt.xticks(seasonal_runs['season_id_x'], rotation=45)
plt.savefig('IPL Total Runs Evolution (Season by Season).jpeg');
plt.show()
```

```python
#Player Performance Across Seasons
def compare_players(player_names):
    player_data = df[df['batter'].isin(player_names)]
    seasonal_player_performance = player_data.groupby(['season_id_x', 'batter'])['batter_runs'].sum().reset_index()
    plt.figure(figsize=(14, 7))
    unique_players = seasonal_player_performance['batter'].unique()
    palette = sns.color_palette("tab10", len(unique_players))
    player_colors = dict(zip(unique_players, palette))
    ax = sns.lineplot(data=seasonal_player_performance, x='season_id_x', y='batter_runs', 
                      hue='batter', palette=player_colors, marker='s', markersize=9, linewidth=2.5)
    for i in range(seasonal_player_performance.shape[0]):
        plt.text(
            x = seasonal_player_performance.season_id_x.iloc[i], 
            y = seasonal_player_performance.batter_runs.iloc[i] + 15, # Offset above the marker
            s = f"{int(seasonal_player_performance.batter_runs.iloc[i])}", 
            ha = 'center', 
            va = 'bottom',
            fontsize = 9,
            fontweight = 'bold'
        )
    all_seasons = sorted(seasonal_player_performance['season_id_x'].unique())
    plt.xticks(all_seasons, rotation=45)

    plt.title('Player Comparison: Runs per Season', fontsize=16)
    plt.ylabel('Runs')
    plt.xlabel('season')
    plt.legend(title='Player')
    plt.xticks(rotation=45)
    plt.savefig('Player Comparison Runs per Season.jpeg');
    plt.show()
compare_players(['V Kohli', 'MS Dhoni',])
```

```python
# Creating a brief summary report
summary = pd.DataFrame({
    'Metric': ['Total Matches Analysed', 'All-time Leading Run Scorer', 'Highest Team Win Rate ID'],
    'Value': [
        len(matches),
        top_10_scorers.iloc[0]['batter'],
        f"ID: {win_rate.index[0]} ({win_rate.values[0]:.2f}%)"
    ]
})

# Exporting summary to CSV
summary.to_csv('ipl_analysis_summary.csv', index=False)
print("Summary insights exported to 'ipl_analysis_summary.csv'")
```
