```python
#importing the csv and extracting first 3 row's
raw_data =np.genfromtxt("Blood_Cancer.csv",
                        delimiter=",",
                        dtype=str,skip_header=1)
#print("header row",raw_data)
print(raw_data.shape,raw_data[:5])
```

```python
#extracting the non categotical (numerical) colums 
num_col =[0,5,9]
numbers_data=raw_data[:,num_col]
print("numerical data:",numbers_data.shape,numbers_data[:5])
```

```python
#converting the numbrs into float valuse using for loop
numbers_data = []

for i in range(raw_data.shape[1]):
    try:
        raw_data[:, i].astype(float)
        numbers_data.append(i)
    except:
        pass

print("Numeric column indices:", numbers_data)
```

```python
numbers_data = raw_data[:, numbers_data].astype(float)
```

```python
#Statistical Operations 
mean_vals = np.mean(numbers_data, axis=0)
np.set_printoptions(precision=2, suppress=True)
print("Mean values per column:\n", mean_vals)

median_vals = np.median(numbers_data, axis=0)
print("Median values per column:\n", median_vals)

std_vals = np.std(numbers_data, axis=0)
print("Standard deviation per column:\n", std_vals)

min_vals = np.min(numbers_data, axis=0)
max_vals = np.max(numbers_data, axis=0)
print("Minimum per column:\n", min_vals)
print("Maximum per column:\n", max_vals)

sum_vals = np.sum(numbers_data, axis=0)
print("Sum per column:\n", sum_vals)

percentiles_25 = np.percentile(numbers_data, 25, axis=0)
percentiles_50 = np.percentile(numbers_data, 50, axis=0)
percentiles_75 = np.percentile(numbers_data, 75, axis=0)
print("25th percentile:\n", percentiles_25)
print("50th percentile (median):\n", percentiles_50)
print("75th percentile:\n", percentiles_75)

corr_matrix = np.corrcoef(numbers_data.T)
print("Correlation matrix:\n", corr_matrix)
```

```python
#categorical analysis
#average Age,WBC,Platelets count by cancer subtype:
cancer_types = raw_data[:, 2] 
unique_types = np.unique(cancer_types)

for ctype in unique_types:
    mask = cancer_types == ctype
    group_values = numbers_data[mask]
    print(ctype, np.mean(group_values, axis=0))
```

```python
#categorical analysis
#average Age,WBC,Platelets count by gender subtype:
cancer_types = raw_data[:, 1] 
unique_types = np.unique(cancer_types)

for ctype in unique_types:
    mask = cancer_types == ctype
    group_values = numbers_data[mask]
    print(ctype, np.mean(group_values, axis=0))
```

```python
#categorical analysis
#average Age,WBC,Platelets count by Side Effects subtype:
cancer_types = raw_data[:, 12] 
unique_types = np.unique(cancer_types)

for ctype in unique_types:
    mask = cancer_types == ctype
    group_values = numbers_data[mask]
    print(ctype, np.mean(group_values, axis=0))
```

```python
#Normalizing the data
max_vals = np.max(numbers_data, axis=1).reshape(-1, 1)
normalized = numbers_data / max_vals
print("Normalized shape:", normalized.shape)
```
